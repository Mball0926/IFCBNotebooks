```python
%matplotlib inline

import numpy as np
from pandas import Series, DataFrame
import pandas as pd
import datetime as dt

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import requests
import json
import re
```

```python
r=requests.get('http://ifcb-data.whoi.edu/mvco/api/feed/temperature/start/2016-06-01/end/2016-07-02')
feed=r.json()
fd=pd.DataFrame(feed)
Py=pd.to_datetime(fd.date)
fd=fd.set_index(Py,drop=False)
fd = fd.drop('date', 1)
fd=fd.drop('temperature',1)
fd['pid']=fd['pid']+'_class_scores.csv'
fd["Cerataulina Count"] = ""
NUM=len(fd.index)
```

```python
k=0
for n in range(0,200):
    dataf=pd.read_csv(fd['pid'][n],index_col='pid')
    dataf.columns
    DF=dataf.idxmax(axis='columns', skipna=True)
    DF.index
    DDF=pd.DataFrame(DF)
    DDF.columns = ['Classification']#rename column
    Count=(DDF=="Cerataulina").sum()
    COUNT=Count.Classification
    fd['Cerataulina Count'][n]=COUNT
    k=k+1
```

```python
px=fd['Cerataulina Count'][0:200]
px.plot(label='Cerataulina Count')
plt.legend()
plt.show()
```

```python
import pandas as pd

PID = 'http://ifcb-data.whoi.edu/mvco/IFCB5_2011_071_012941'
FEATURES_URL = PID + '_features.csv'

df = pd.read_csv(FEATURES_URL)
df.head()
```

```python
s = df['Area'] / df['ConvexArea']
s.sort_values(inplace=True)

con = df.ix[s.index]
con.head()
```

```python
roi_number = con['roi_number'].iloc[0]
roi_number
```

```python
fmt = '%s_%05d'

ROI_PID = fmt % (PID, roi_number)

ROI_PID
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

from skimage.io import imread

img = imread(ROI_PID + '.png')

plt.figure(figsize=(15,15))
plt.imshow(img, cmap='gray', vmin=0, vmax=255, interpolation='nearest')
```
